```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
from xgboost import XGBRegressor
import joblib
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
#Missing Values
missing_values = df.isna().sum().sort_values(ascending = False)
missing_values
```

```python
missing_values_plot = missing_values[missing_values > 500]
plt.figure(figsize=(10,5))
sns.barplot(x=missing_values_plot,y=missing_values_plot.index, palette='viridis')
plt.title("Missing Values by Feature")
plt.xlabel("Missing Values")
plt.ylabel("Feature")
plt.show()
```

```python
df.head()
```

```python
df = df.drop(columns=["Id", "PoolQC", "MiscFeature", "Alley", "Fence", "MasVnrType", "FireplaceQu"])
df
```

```python
df.columns
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)
```

```python
label_encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

df
```

```python
corr_matrix = df.corr()
corr_with_price = corr_matrix['SalePrice'].sort_values(ascending=False)
print("Correlated Features", corr_with_price.head(20))
```

```python
c_features = corr_with_price[corr_with_price>0.25].index
df = df[c_features]
df
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df[corr_with_price[corr_with_price>0.25].index].corr(),annot=True, fmt='.2f',cmap='Reds')
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
df = df.drop(columns=['GarageCars','GarageYrBlt','TotRmsAbvGrd','Foundation','2ndFlrSF','HalfBath','TotalBsmtSF'])
df
```

```python
df.isna().sum()
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
df.isna().sum()
```

```python
x = df.drop("SalePrice", axis =1)
Y = df["SalePrice"]
```

```python
x_train, x_test, Y_train, Y_test = train_test_split(x, Y, train_size=0.2, random_state=42)
```

```python
x
```

```python
Y
```

```python
model = XGBRegressor(
    objective = 'reg:squarederror',
    learning_rate = 0.01,
    max_depth = 6,
    n_estimators = 1000,
    subsample = 0.65,
    random_state = 42
)
model.fit(x_train,Y_train)
```

```python
Y_pred = model.predict(x_test)
```

```python
mae = mean_absolute_error(Y_test, Y_pred)
print("Mean Absolute Error: ", mae)
```

```python
r2_score = r2_score(Y_test,Y_pred)
print("r2 score:", r2_score) 
```

```python
joblib.dump(model,"xgb_model.jb")
```
